## 3. ***Data Wrangling***

### Data Wrangling Code

````python
import pandas as pd
import numpy as np
import re


# Handle missing values
df2['age_certification'] = df2['age_certification'].fillna('Unknown')
df2['runtime'] = df2['runtime'].fillna(df2['runtime'].median())
df2['genres'] = df2['genres'].fillna('Unknown')
df2['imdb_score']= df2['imdb_score'].fillna(0)
df2['tmdb_popularity']= df2['tmdb_popularity'].fillna(0)
df2['tmdb_score']= df2['tmdb_score'].fillna(0)
df2['imdb_votes']= df2['imdb_votes'].fillna(0)


# Drop rows in credits where 'name' is missing
df1 = df1.dropna(subset=['name'])

#Convert unhashables columns (like lists) to strings
def stringify_unhashables(df):
    for col in df.columns:
        if df[col].apply(lambda x: isinstance(x, (list, dict))).any():
            df[col] = df[col].apply(str)
    return df
df1 = stringify_unhashables(df1)
df2 = stringify_unhashables(df2)

key_columns = ['id']
duplicates = df1[df1.duplicated(subset=key_columns, keep=False)]
print(duplicates.sort_values(by=key_columns))

key_columns = ['id']
duplicates = df2[df2.duplicated(subset=key_columns, keep=False)]
print(duplicates.sort_values(by=key_columns))


# Remove duplicates
df2 = df2.drop_duplicates(subset=key_columns)
df1 = df1.drop_duplicates(subset=key_columns)


# Convert data types
df2['release_year'] = df2['release_year'].astype('Int64')
df2['runtime'] = df2['runtime'].astype('float')

# Split genres into lists
df2['genres'] = df2['genres'].apply(lambda x: x.split(', ') if isinstance(x, str) else [])

# Merge datasets
merged_df = pd.merge(df2, df1, on='id', how='inner')

# Split (explode) the 'genres' and 'production_countries' columns so that each genre and country has its own row.
df3final=merged_df.explode("genres").explode("production_countries").reset_index(drop=True)

#Fill missing values of final dataset
df3final.fillna("Not available",inplace=True)


# Clean up deeply nested genre strings
def extract_genres(raw):
    # Extract quoted words like "comedy", 'drama' etc
    return re.findall(r'\b[a-zA-Z][a-zA-Z\s&-]*\b', raw)

# Apply the cleaning
df3final['genres_cleaned'] = df3final['genres'].astype(str).apply(extract_genres)

# Reset index (for clean dataframes)
df2.reset_index(drop=True, inplace=True)
df1.reset_index(drop=True, inplace=True)
merged_df.reset_index(drop=True, inplace=True)
merged_df = stringify_unhashables(merged_df)
merged_df = merged_df.drop_duplicates().reset_index(drop=True)
merged_df = merged_df.drop_duplicates(subset=key_columns)

# 10. Save cleaned data
df2.to_csv('df2_clean.csv', index=False)
df1.to_csv('df1_clean.csv', index=False)
merged_df.to_csv('merged_clean.csv', index=False)
df3final.to_csv('df3final_clean.csv', index=False)

print("\nNow the dataset is ready for Data Analysis.")
````

* **Missing Values Handled**:

  * Filled missing `age_certification` and `genres` with `'Unknown'`.
  * Filled missing `runtime` with the median value.
  * Dropped rows in `df1` where `name` was missing.

* **Unhashable Data Fixed**:

  * Converted list/dict-type columns to strings for compatibility.

* **Duplicates Removed**:

  * Identified and dropped duplicate rows based on the `id` column in both datasets.

* **Data Types Standardized**:

  * Converted `release_year` to integer and `runtime` to float.

* **Genres Column Normalized**:

  * Split comma-separated genre strings into lists.

* **Datasets Merged**:

  * Merged `df1` and `df2` on `id` to create a unified dataset.

* **Final Cleanup**:

  * Reset indices and ensured no duplicate rows remain.
  * Saved all cleaned datasets to CSV files.
